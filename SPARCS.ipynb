```python
%matplotlib inline
from matplotlib import pylab as plt
import pandas as pd
df = pd.read_csv("2012/Hospital_Inpatient_Discharges_SPARCS_De-Identified_2012.csv")
```

```python
df.columns
```

```python
import string
tbl = string.maketrans("", "")

df['Total Charges'] = df['Total Charges'].apply(lambda x: float(x.translate(tbl, ",")))
```

```python
df['Total Charges'].hist(bins=200, figsize=(16., 10.))
```

```python
df[df['Total Charges'] < 250000]['Total Charges'].hist(bins=200 ,figsize=(16., 10.))
```
